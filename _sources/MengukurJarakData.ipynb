Penjelasan :
Kode tersebut digunakan untuk menghitung jarak antara dua data setelah seluruh variabel kategorikal diubah menjadi bentuk numerik menggunakan teknik one-hot encoding. Pendekatan ini memungkinkan perhitungan jarak Euclidean dan Manhattan dilakukan tanpa memisahkan tipe data seperti pada metode Gower.

Pertama, dataset diubah menggunakan fungsi get_dummies. Proses ini mengubah semua variabel kategorikal menjadi beberapa kolom biner bernilai nol dan satu. Misalnya, jika sebuah kolom memiliki tiga kategori, maka akan dibuat beberapa kolom baru yang menunjukkan keberadaan masing-masing kategori. Opsi drop_first bernilai true digunakan untuk menghindari duplikasi informasi yang tidak perlu.

Setelah itu, seluruh data dikonversi menjadi tipe float agar bisa dihitung secara matematis. Hal ini penting karena perhitungan jarak dengan numpy hanya dapat dilakukan pada data numerik.

Kemudian, dua baris pertama dari dataset yang sudah diubah diambil sebagai vektor p dan q. Kedua vektor ini berisi seluruh nilai atribut dalam bentuk angka sehingga bisa langsung dikurangkan satu sama lain.

Selanjutnya dihitung dua jenis jarak. Jarak Euclidean dihitung menggunakan fungsi norm dari numpy, yang secara konsep mengukur jarak garis lurus antara dua titik dalam ruang multidimensi. Jarak Manhattan dihitung dengan menjumlahkan selisih absolut dari setiap atribut, yang secara konsep mengukur jarak seperti pergerakan pada grid.

Terakhir, hasil kedua jarak tersebut ditampilkan. Nilai yang lebih kecil menunjukkan bahwa kedua data lebih mirip, sedangkan nilai yang lebih besar menunjukkan perbedaan yang semakin jauh.

```python
import numpy as np
import pandas as pd

# Ambil 2 data
data_1 = df.iloc[0]
data_2 = df.iloc[1]

# ==============================
# KELOMPOK TIPE DATA
# ==============================

kolom_numerik = ['age','absences','G1','G2','G3']

kolom_ordinal = [
    'traveltime','studytime','failures',
    'famrel','freetime','goout',
    'Dalc','Walc','health'
]

kolom_nominal = [
    'school','sex','address','famsize','Pstatus',
    'Mjob','Fjob','reason','guardian'
]

kolom_binary = [
    'schoolsup','famsup','paid','activities',
    'nursery','higher','internet','romantic'
]

# ==============================
# VARIABEL RUMUS GOWER
# ==============================

jumlah_jarak = 0.0     # Σ(δ * d)
jumlah_delta = 0       # Σδ

# ==============================
# NOMINAL
# ==============================

for kolom in kolom_nominal:
    if pd.notna(data_1[kolom]) and pd.notna(data_2[kolom]):
        jarak = 0 if data_1[kolom] == data_2[kolom] else 1
        jumlah_jarak += jarak
        jumlah_delta += 1

# ==============================
# BINARY SIMETRIS
# ==============================

for kolom in kolom_binary:
    if pd.notna(data_1[kolom]) and pd.notna(data_2[kolom]):
        nilai_1 = 1 if str(data_1[kolom]).lower() == 'yes' else 0
        nilai_2 = 1 if str(data_2[kolom]).lower() == 'yes' else 0
        
        jarak = 0 if nilai_1 == nilai_2 else 1
        jumlah_jarak += jarak
        jumlah_delta += 1

# ==============================
# ORDINAL
# ==============================

for kolom in kolom_ordinal:
    if pd.notna(data_1[kolom]) and pd.notna(data_2[kolom]):
        
        min_val = df[kolom].min()
        max_val = df[kolom].max()
        
        if max_val != min_val:
            z1 = (data_1[kolom] - min_val) / (max_val - min_val)
            z2 = (data_2[kolom] - min_val) / (max_val - min_val)
            
            jarak = abs(z1 - z2)
            jumlah_jarak += jarak
            jumlah_delta += 1

# ==============================
# NUMERIK
# ==============================

for kolom in kolom_numerik:
    if pd.notna(data_1[kolom]) and pd.notna(data_2[kolom]):
        
        min_val = df[kolom].min()
        max_val = df[kolom].max()
        
        if max_val != min_val:
            jarak = abs(data_1[kolom] - data_2[kolom]) / (max_val - min_val)
            jumlah_jarak += jarak
            jumlah_delta += 1

# ==============================
# HASIL AKHIR
# ==============================

jarak_campuran = jumlah_jarak / jumlah_delta

print("Hasil Perhitungan Jarak Data Campuran")
print("-------------------------------------")
print("Total penjumlahan seluruh nilai jarak tiap atribut =", jumlah_jarak)
print("Jumlah atribut yang ikut dihitung =", jumlah_delta)
print("Nilai akhir jarak campuran antara data 1 dan data 2 =", jarak_campuran)

if jarak_campuran < 0.3:
    print("Interpretasi: Kedua data sangat mirip.")
elif jarak_campuran < 0.6:
    print("Interpretasi: Kedua data cukup mirip.")
else:
    print("Interpretasi: Kedua data berbeda cukup jauh.")
```

Penjelasan :
Kode tersebut digunakan untuk menghitung jarak antara dua data pada dataset yang memiliki tipe atribut campuran, yaitu numerik, ordinal, nominal, dan biner. Perhitungan ini mengikuti konsep Gower Distance, yaitu metode yang memang dirancang untuk menghitung jarak pada data dengan berbagai tipe sekaligus dalam satu perhitungan.

Pertama, program mengambil dua baris data dari dataframe. Kedua data ini kemudian disimpan dalam variabel data satu dan data dua. Semua perhitungan selanjutnya adalah proses membandingkan setiap atribut dari kedua data tersebut.

Selanjutnya, kolom-kolom pada dataset dikelompokkan berdasarkan jenis tipenya. Kolom numerik berisi nilai angka murni seperti umur, jumlah absen, dan nilai ujian. Kolom ordinal berisi angka yang memiliki tingkatan atau urutan, seperti waktu belajar atau tingkat kesehatan. Kolom nominal berisi kategori tanpa urutan seperti jenis sekolah atau jenis kelamin. Sedangkan kolom biner berisi dua kemungkinan nilai seperti yes dan no.

Program kemudian menghitung jarak untuk setiap tipe data dengan cara yang berbeda. Untuk data nominal, jika nilainya sama maka jaraknya nol, jika berbeda maka jaraknya satu. Untuk data biner simetris, nilai yes diubah menjadi satu dan no menjadi nol, lalu dibandingkan dengan cara yang sama seperti nominal. Untuk data ordinal, nilainya terlebih dahulu dinormalisasi ke skala nol sampai satu berdasarkan nilai minimum dan maksimum pada kolom tersebut, lalu dihitung selisih absolutnya. Untuk data numerik, selisih absolut antara kedua nilai dibagi dengan rentang nilai kolom agar berada pada skala nol sampai satu.

Setiap hasil jarak dari satu atribut akan ditambahkan ke variabel jumlah jarak, dan setiap atribut yang berhasil dihitung akan menambah jumlah atribut yang dibandingkan. Setelah semua atribut selesai diproses, nilai akhir jarak campuran diperoleh dengan membagi total jarak dengan jumlah atribut yang dihitung. Nilai ini berada pada rentang nol sampai satu, di mana nilai mendekati nol berarti kedua data sangat mirip, sedangkan nilai mendekati satu berarti kedua data semakin berbeda.
