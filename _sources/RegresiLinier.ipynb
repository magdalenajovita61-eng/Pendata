## 3. Perhitungan Regresi Linier secara Analitik

Persamaan pada regresi linier umumnya dinyatakan dalam bentuk:

image.png

### Keterangan:

Persamaan regresi yang diperoleh adalah:

$$
y = 1.05x + 0.4
$$

Keterangan:

- $y$ = variabel terikat (nilai prediksi)
- $x$ = variabel bebas
- $a = 0.4$ = konstanta (intercept), yaitu nilai $y$ ketika $x = 0$
- $b = 1.05$ = koefisien regresi (slope), yaitu perubahan nilai $y$ untuk setiap kenaikan 1 satuan pada $x$

Berdasarkan persamaan tersebut, setiap kenaikan nilai $x$ sebesar 1 satuan akan meningkatkan nilai $y$ sebesar 1.05.

### Tabel Perhitungannya

| x | y | x² | xy |
|---|---|---|---|
| 2 | 2 | 4  | 4  |
| 4 | 3 | 16 | 12 |
| 3 | 5 | 9  | 15 |
| 3 | 4 | 9  | 12 |
| 3 | 3 | 9  | 9  |
| 4 | 5 | 16 | 20 |
| 5 | 6 | 25 | 30 |
| **24** | **28** | **88** | **102** |


- Menghitung Koefisien Regresi

$$
b = \frac{n\sum xy - (\sum x)(\sum y)}{n\sum x^2 - (\sum x)^2}
$$

$$
b = \frac{7(102) - (24)(28)}{7(88) - (24)^2}
$$

$$
b = \frac{714 - 672}{616 - 576}
$$

$$
b = \frac{42}{40}
$$

$$
b = 1.05
$$

- Menghitung Konstanta

$$
a = \frac{\sum y - b\sum x}{n}
$$

$$
a = \frac{28 - (1.05)(24)}{7}
$$

$$
a = \frac{28 - 25.2}{7}
$$

$$
a = \frac{2.8}{7}
$$

$$
a = 0.4
$$

- Persamaan Regresi

$$
y = a + bx
$$

$$
y = 0.4 + 1.05x
$$

## 4. Persamaan Regresi

Berdasarkan hasil perhitungan regresi linier, diperoleh persamaan regresi sebagai berikut:

$$
y = 0.4 + 1.05x
$$

Persamaan tersebut menunjukkan hubungan antara variabel $x$ dan variabel $y$. Nilai koefisien regresi sebesar 1.05 menandakan bahwa setiap kenaikan nilai $x$ sebesar 1 satuan akan menyebabkan nilai $y$ meningkat sebesar 1.05.

## 5. Implementasi Menggunakan Python dan Sklearn

Berikut kode program menggunakan library sklearn:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([2,4,3,3,3,4,5]).reshape(-1,1)
y = np.array([2,3,5,4,3,5,6])

model = LinearRegression()
model.fit(x, y)

print("Koefisien Regresi :", model.coef_[0])
print("Intercept :", model.intercept_)
```

### Hasil Output

```python
Koefisien Regresi : 1.05
Intercept : 0.4
```

### Penjelasan Setiap Kode

```python
from sklearn.linear_model import LinearRegression
```

Kode di atas digunakan untuk mengimpor class `LinearRegression` dari library `sklearn` yang berfungsi untuk membuat model regresi linier.


```python
import numpy as np
```

Kode ini digunakan untuk mengimpor library `numpy` dengan nama singkat `np`. Library ini digunakan untuk mengelola dan menghitung data berbentuk array.

```python
x = np.array([2,4,3,3,3,4,5]).reshape(-1,1)
```

Kode tersebut digunakan untuk membuat data variabel `x` dalam bentuk array.

Kode ini digunakan untuk membuat data variabel `y` sebagai target atau nilai yang akan diprediksi.

```python
model = LinearRegression()
```

Kode tersebut digunakan untuk membuat objek model regresi linier.

```python
model.fit(x, y)
```

Kode ini digunakan untuk melatih model regresi menggunakan data `x` dan `y` sehingga model dapat menemukan garis regresi terbaik.

```python
print("Koefisien :", model.coef_[0])
```

Kode tersebut digunakan untuk menampilkan nilai koefisien regresi atau slope dari garis regresi.

```python
print("Intercept :", model.intercept_)
```
