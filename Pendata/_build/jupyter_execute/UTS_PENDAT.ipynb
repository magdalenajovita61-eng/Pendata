# Klasifikasi Kesuburan Tanah Menggunakan K-Nearest Neighbors (KNN)



# Metodologi dan Hasil Analisis

## 1. Data Preprocessing

Tahapan preprocessing dilakukan untuk mempersiapkan data sebelum digunakan dalam pemodelan:

- **Handling Missing Value**  
  Data yang memiliki nilai kosong ditangani menggunakan node *Missing Value* dengan metode pengisian yang sesuai untuk setiap tipe data.

- **Encoding Data Kategorikal**  
  Kolom kategorikal seperti *Tekstur Tanah* diubah menjadi numerik menggunakan metode **One to Many (One-Hot Encoding)** agar dapat diproses oleh algoritma.

- **Normalisasi Data**  
  Seluruh fitur numerik dinormalisasi menggunakan metode **Min-Max (0–1)**.  
  Hal ini dilakukan karena algoritma KNN berbasis jarak sehingga membutuhkan skala data yang seragam.

## 2. Pembagian Data

Dataset dibagi menjadi dua bagian:

- **Data Training: 70%**
- **Data Testing: 30%**

Metode pembagian menggunakan **Stratified Sampling**, sehingga proporsi kelas tetap seimbang antara data training dan testing.


## 3. Model K-Nearest Neighbors (KNN)

Algoritma yang digunakan adalah **K-Nearest Neighbors (KNN)** dengan parameter:

- Nilai K = **5**
- Menggunakan perhitungan jarak (Euclidean distance)

Model bekerja dengan cara:
- Menghitung jarak antara data uji dan data training
- Mengambil K tetangga terdekat
- Menentukan kelas berdasarkan mayoritas tetangga

## 4. Evaluasi Model

Evaluasi dilakukan menggunakan node **Scorer** dengan membandingkan:

- Kolom aktual: **Label**
- Kolom prediksi: **Class (kNN)**

Hasil ini menunjukkan bahwa data memiliki pola yang sangat jelas dan dapat dipisahkan dengan baik oleh algoritma KNN.

## 7. Perhitungan Evaluasi

Berdasarkan confusion matrix:

| Actual \ Predicted | Tidak Subur | Subur |
|------------------|------------|------|
| Tidak Subur      | 300        | 0    |
| Subur            | 0          | 300  |

Maka diperoleh:

- **TP (True Positive) = 300**
- **TN (True Negative) = 300**
- **FP (False Positive) = 0**
- **FN (False Negative) = 0**

---

### 1. Accuracy

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

$$
Accuracy = \frac{300 + 300}{300 + 300 + 0 + 0}
$$

$$
Accuracy = \frac{600}{600} = 1 = 100\%
$$

---

### 2. Precision

$$
Precision = \frac{TP}{TP + FP}
$$

$$
Precision = \frac{300}{300 + 0}
$$

$$
Precision = \frac{300}{300} = 1 = 100\%
$$

---

### 3. Recall

$$
Recall = \frac{TP}{TP + FN}
$$

$$
Recall = \frac{300}{300 + 0}
$$

$$
Recall = \frac{300}{300} = 1 = 100\%
$$

---

### 4. F1-Score

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

$$
F1 = 2 \times \frac{1 \times 1}{1 + 1}
$$

$$
F1 = 2 \times \frac{1}{2} = 1 = 100\%
$$

## 8. Kesimpulan

Berdasarkan hasil evaluasi menggunakan algoritma K-Nearest Neighbors (KNN), diperoleh nilai Accuracy, Precision, Recall, dan F1-Score masing-masing sebesar **100%**.

Hal ini menunjukkan bahwa model mampu mengklasifikasikan seluruh data uji dengan benar tanpa kesalahan. Tidak terdapat nilai False Positive (FP) maupun False Negative (FN), sehingga semua prediksi sesuai dengan kondisi sebenarnya.

Performa model yang sangat tinggi ini mengindikasikan bahwa data memiliki pola yang jelas dan fitur yang digunakan sangat relevan dalam membedakan kelas **Subur** dan **Tidak Subur**.

Namun demikian, hasil yang sempurna juga perlu diperhatikan karena dapat menunjukkan bahwa dataset relatif mudah dipisahkan atau berpotensi mengalami overfitting. Oleh karena itu, disarankan untuk melakukan pengujian lebih lanjut dengan variasi parameter atau dataset yang berbeda untuk memastikan kemampuan generalisasi model.

Secara keseluruhan, algoritma KNN terbukti efektif dalam menyelesaikan permasalahan klasifikasi kesuburan tanah pada dataset ini.
